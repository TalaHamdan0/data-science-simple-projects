```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'title': ['the dark knight', 'inception', 'interstellar', 'parasite', 'joker',
              'avengers: endgame', 'titanic', 'the godfather', 'forrest gump', 'gladiator'],
    'year': [2008, 2010, 2014, 2019, 2019, 2019, 1997, 1972, 1994, 2000],
    'genre': ['action', 'sci-fi', 'sci-fi', 'drama', 'crime',
              'action', 'romance', 'crime', 'drama', 'action'],
    'rating': [9.0, 8.8, 8.6, 8.6, 8.5, 8.4, 7.9, 9.2, 8.8, 8.5],
    'revenue_million': [1005, 829, 677, 258, 1074, 2797, 2187, 245, 678, 460]
}

df = pd.DataFrame(data)

print('Top 5 movies by rating:')
print(df.sort_values('rating', ascending=False).head(5))

print('\nMovies production count by year:')
print(df['year'].value_counts())

print('\nAverage revenue (in million):')
print(df['revenue_million'].mean())

avg_revenue_by_genre = df.groupby("genre")["revenue_million"].mean()
print("avg_revenue_by_genre is :", avg_revenue_by_genre)  
plt.figure(figsize=(8,5))
avg_revenue_by_genre.plot(kind="bar",color="teal")
plt.title("avg genre")
plt.xlabel('genre')
plt.ylabel('million')
plt.show()

#scatter
plt.figure(figsize=(8,5))
plt.scatter(df['rating'],df['revenue_million'],color="purple",alpha=0.7)
plt.title("relation batween rating and revenue million")
plt.xlabel("rating")
plt.ylabel("revenue_million")
plt.grid(True)
plt.show()
most_comon_genre= df['genre'].value_counts().idxmax()
print("the most gnre is:",most_comon_genre)

#pie 
gnre_counts=df['genre'].value_counts()
plt.figure(figsize=(8,5))
gnre_counts.plot(kind="pie",autopct="%1.1f%%",color=["pink","green","blue","black","brown"])
plt.title("type of movies")
plt.xlabel("movie")
plt.ylabel(" ")
plt.show()


movies_per_year = df['year'].value_counts().sort_index()
#bar
plt.figure(figsize=(8, 5))
movies_per_year.plot(kind="bar", color="skyblue")
plt.title("Number of Movies per Year")
plt.xlabel("Year")
plt.ylabel("Number of Movies")
plt.show()

```
